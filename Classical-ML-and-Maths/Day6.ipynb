# 3. Hands-on
## Use numpy to implement vector & matrix operations

```python
import numpy as np
# Create matrices
a = np.array([[2, 4], [1, 5]])
b = np.array([[6, 8], [7, 9]])

# Addition
print("Matrix Addition:\n", a + b)

# Subtraction
print("Matrix Subtraction:\n", a- b)

# Matrix multiplication
print("Matrix Multiplication:\n", np.dot(a, b))

# Element-wise multiplication
print("Element-wise Multiplication:\n", a * b)

# Transpose
print("Transpose of a:\n", a.T)

# Determinant
print("Determinant of a:", np.linalg.det(a))
```

## Plot probability distributions with matplotlib

```python
import matplotlib.pyplot as plt

# 1. Normal Distribution
mu, sigma = 0, 1
x = np.linspace(-5, 5, 500)
normal_pdf = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# 2. Uniform Distribution
a, b = 0, 1
uniform_pdf = np.ones_like(x) / (b - a)

# 3. Binomial Distribution
n, p = 10, 0.5
binomial_pmf = np.random.binomial(n, p, size=10000)

# 4. Exponential Distribution
lam = 1
exp_pdf = lam * np.exp(-lam * x)

# Plot all
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, normal_pdf, label="Normal(Î¼=0, Ïƒ=1)", color='blue')
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, uniform_pdf, label="Uniform(0,1)", color='green')
plt.title("Uniform Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.ylim(0, 2)
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(binomial_pmf, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.7, color='orange')
plt.title(f"Binomial(n={n}, p={p})")
plt.xlabel("k")
plt.ylabel("Probability")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(x[x >= 0], exp_pdf[x >= 0], label="Exponential(Î»=1)", color='red')
plt.title("Exponential Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# 4.Regression
## Linear Regression:
### It is a statistical model that analyzes the linear relationship between a dependent variable and a given set of independent variables.
### In ML, linear regression is used for predicting continuous numeric values based on learned linear relation for new and unseen data. It is used in predictive modeling, financial forecasting, risk assessment, etc. 
## Ordinary Least Square (OLS): 
### It is  a widely used statistical method for estimating the parameters of a linear regression model that minimizes the sum of squared residuals between observed and predicted values.


### Implementing Ordinary Least Squares (OLS) linear regression using scikit-learn


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
d = pd.read_csv("Breast_cancer_dataset.csv")
d.head()
```

```python
x = d[['radius_mean']]
y= d['area_worst']
model = LinearRegression()  # Create a linear regression model
model.fit(x,y)  # Train the model to find the best fit line predicting y from x
print("Intercept:", model.intercept_) # Print the intercept (beta0)
print("Coefficient:", model.coef_[0]) # Print the slope (beta1) for radius_mean
```

```python
y_pred = model.predict(x)
plt.scatter(x, y, color='Purple', label='Actual')
plt.plot(x, y_pred, color='pink', label='Predicted')
plt.xlabel('radius_mean')
plt.ylabel('area_worst')
plt.legend()
plt.show()
```
