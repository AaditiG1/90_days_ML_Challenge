# 1. Linear Algebra
## Vectors
### It is something that has both magnitude and direction.

## Dot-Product
### It is the multiplication between two vectors.

#### Let 𝒗=(𝑣1,𝑣2,𝑣3) and 𝒘 = (𝑤1,𝑤2,𝑤3) be vectors in ℝ3. The dot product of 𝒗 and 𝒘, denoted by 𝒗 ∙ 𝒘, is given by: 
#### 𝒗∙𝒘=𝑣1𝑤1+𝑣2𝑤2+𝑣3𝑤3 Similarly, for vectors 𝒗 = (𝑣1,𝑣2)  and 𝒘 = (𝑤1,𝑤2) in ℝ2, the dot product is: 𝒗∙𝒘=𝑣1𝑤1+𝑣2𝑤2

## Matrix Multiplication:
### It is a fundamental operation in linear algebra where you multiply two matrices to produce a third matrix.
### Rule: 
#### 1. You can only multiply matrix A by matrix B if the number of columns in A equals the number of rows in B
#### 2. If A is m×n and B is n×p, the result C will be m×p

## Matrix Transpose:   
### -> rows become columns and columns become rows.

## Matrix Inverse:  
### -> Only square matrices can have inverses.

## Eigenvalues 
### -> Indicate how much an eigenvector gets stretched or compressed during the transformation. The eigenvector's direction remains unchanged unless the eigenvalue is negative, in which case the direction is simply reversed.

## Eigenvectors
### They are non-zero vectors that, when multiplied by a matrix, only stretch or shrink without changing direction. The eigenvalue must be found first before the eigenvector. 

#### Equation: |A−λI| = 0

# 2. Probability and Stats

## Standard Deviation: It indicates the extent of the data's variation and shows how far individual data points deviate from the average. 

## Variance: It is a statistical measure that tells us how spread out the values in a data set are from the mean (average).

## Conditional Probability: 
### The probability of an event based on the fact that some other event has occurred, will occur, or is occurring is conditional probability.
#### p(A/B) = P(A and B) / P(B) -> for the conditional probability of A given B has occurred
#### p(B/A) = P(B and A) / P(A) -> for the conditional probability of B given A has occurred

### P(B/A) P(A) / P(B) - for two events A and B

# 3. Hands-on
## Use numpy to implement vector & matrix operations

```python
import numpy as np
# Create matrices
a = np.array([[2, 4], [1, 5]])
b = np.array([[6, 8], [7, 9]])

# Addition
print("Matrix Addition:\n", a + b)

# Subtraction
print("Matrix Subtraction:\n", a- b)

# Matrix multiplication
print("Matrix Multiplication:\n", np.dot(a, b))

# Element-wise multiplication
print("Element-wise Multiplication:\n", a * b)

# Transpose
print("Transpose of a:\n", a.T)

# Determinant
print("Determinant of a:", np.linalg.det(a))
```

## Plot probability distributions with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

# Create data for a normal distribution
mean = 0      # Center of the curve
std_dev = 1   # Standard deviation
x = np.linspace(-1, 6, 1000)  # Values from -4 to 4
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2) # Gaussian distribution 
# Plot
plt.plot(x, y, label='Normal Distribution', color='maroon')
plt.title("Probability Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

# 4.Regression
## Linear Regression:
### It is a statistical model that analyzes the linear relationship between a dependent variable and a given set of independent variables.
### In ML, linear regression is used for predicting continuous numeric values based on learned linear relation for new and unseen data. It is used in predictive modeling, financial forecasting, risk assessment, etc. 
## Ordinary Least Square (OLS): 
### It is  a widely used statistical method for estimating the parameters of a linear regression model that minimizes the sum of squared residuals between observed and predicted values.


### Implementing Ordinary Least Squares (OLS) linear regression using scikit-learn


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
d = pd.read_csv("Breast_cancer_dataset.csv")
d.head()
```

```python
x = d[['radius_mean']]
y= d['area_worst']
model = LinearRegression()  # Create a linear regression model
model.fit(x,y)  # Train the model to find the best fit line predicting y from x
print("Intercept:", model.intercept_) # Print the intercept (beta0)
print("Coefficient:", model.coef_[0]) # Print the slope (beta1) for radius_mean
```

```python
y_pred = model.predict(x)
plt.scatter(x, y, color='Purple', label='Actual')
plt.plot(x, y_pred, color='pink', label='Predicted')
plt.xlabel('radius_mean')
plt.ylabel('area_worst')
plt.legend()
plt.show()
```
