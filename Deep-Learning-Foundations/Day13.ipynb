# 🧠 Multi-Layer Perceptron (MLP)



---

## 🔹 Defination of MLP:

A **Multilayer Perceptron (MLP)** is a type of **artificial neural network** with multiple layers of neurons.

- It has **three main parts**:
    - **Input Layer** → Receives raw data.
    - **Hidden Layers** → Extract patterns/features.
    - **Output Layer** → Produces the final prediction.

It is a **feedforward network** → data flows only in one direction (no loops).

---

## 🔹 Structure of an MLP

**Flow of data:**

Input → Hidden Layer 1 → Hidden Layer 2 → ... → Output

Each hidden layer performs:



---

## 🔹 Working of MLP:

1. **Forward Pass** 🚀
    
    Input passes through layers to produce predictions.
    
2. **Loss Calculation** ⚖️
    
    Prediction is compared with the true label using a loss function (e.g., Cross-Entropy, MSE).
    
3. **Backward Pass (Backpropagation)** 🔄
    
    Gradients are calculated to see how much each weight contributed to the error.
    
4. **Weight Update (Gradient Descent / Optimizers)** 🎯
    
    Weights and biases are updated to reduce errors for the next iteration.
    

---

## 🔹 Why Do Hidden Layers Matter?

- **Without activation functions** → Multiple layers collapse into a single linear transformation → cannot model complex data.
- **With non-linear activations (ReLU, Sigmoid, Tanh)** → The network can model **non-linear decision boundaries**.

👉 Example:

- A single perceptron cannot solve the **XOR problem** (not linearly separable).
- Adding a hidden layer with non-linear activation allows MLP to solve it ✅.

---

## 🔹 Key Points About MLPs

- **Universal Approximator** 🏆
    - With enough neurons, an MLP can approximate any continuous function.
- **Multiple Hidden Layers → Hierarchical Learning**
    - Layer 1: learns **simple features** (edges, small patterns).
    - Layer 2: learns **complex features** (shapes, textures).
    - Higher Layers: capture **abstract concepts** (faces, objects).
- **Applications** 🌍
    - Image classification
    - Speech recognition
    - Natural Language Processing (NLP)
    - Financial forecasting
    - Medical diagnosis
- **Challenges** ⚠️
    - Requires large data & computation.
    - Risk of overfitting.
    - Choosing layers/neurons is tricky.

---

## ✅ Summary

- An **MLP** is a powerful network that learns patterns by stacking layers.
- **Hidden layers + Non-linear activations** give MLPs the ability to solve **non-linear problems**.
- They form the **foundation of deep learning models**.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

# --- Load MNIST dataset ---
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize images
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

print("x_train:", x_train.shape, "y_train:", y_train.shape)
print("x_test:", x_test.shape, "y_test:", y_test.shape)
```

**Code Explination**
- Loads **MNIST dataset** (28x28 grayscale images of digits 0–9).
- Normalizes pixel values to **0–1** by dividing by 255.
- `x_train.shape = (60000,28,28)`, `y_train.shape = (60000,)`.
- `x_test.shape = (10000,28,28)`, `y_test.shape = (10000,)`.

```python
# --- Function to build MLP ---
def build_mlp(use_activation=True):
    activation = 'relu' if use_activation else None
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(256, activation=activation),   # 1st hidden layer
        Dense(128, activation=activation),   # 2nd hidden layer
        Dense(10, activation='softmax')      # output layer
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

**Code Explination**
- `Flatten` → Converts 28×28 images into a 784-element vector (input layer).
- `Dense` → Fully connected layer.
    - **256 neurons**, **128 neurons** → hidden layers
    - `activation='relu'` or `None` → can toggle non-linearity.
- Output layer: 10 neurons with `softmax` → probability for each digit.
- `optimizer='adam'` → adjusts weights to reduce loss.
- `loss='sparse_categorical_crossentropy'` → suitable for multi-class classification.

```python
# --- Train model with activation ---
model_with_act = build_mlp(use_activation=True)
history_with_act = model_with_act.fit(x_train, y_train, 
                                      epochs=10, 
                                      batch_size=2000, 
                                      validation_split=0.2, verbose=1)
# Evaluate
loss_with_act, acc_with_act = model_with_act.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy WITH activation: {acc_with_act*100:.2f}%")
```

**Code Explination**
- Builds a **model with ReLU activations**.
- Trains for 10 epochs, batch size = 2000.
- `validation_split=0.2` → 20% of training data is used to check validation performance.
- Evaluates **test accuracy** of the trained model.

**Output Explination**

```python
# --- Train model without activation ---
model_without_act = build_mlp(use_activation=False)
history_without_act = model_without_act.fit(x_train, y_train, 
                                            epochs=10, 
                                            batch_size=2000, 
                                            validation_split=0.2, verbose=1)

# Evaluate
loss_without_act, acc_without_act = model_without_act.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy WITHOUT activation: {acc_without_act*100:.2f}%")
```

# 🔄 Forward Pass (Forward Propagation) in an MLP



##  Definition & Overview

The **forward pass** (also called **forward propagation**) is the process where input data **moves through a neural network (MLP) layer by layer**, producing intermediate outputs in the hidden layers and a final output at the output layer.

During this process, each layer performs:

1. **Linear Transformation**:
    - Inputs are multiplied by **weights** and a **bias** is added:
      

   
2. **Non-Linear Activation**:
    - The linear combination is passed through a **non-linear activation function** (e.g., ReLU, Sigmoid, Tanh):
   


    - This allows the network to **model complex, non-linear patterns**.

⚡ **Summary**: Forward pass = **Input → Linear Transform → Non-Linearity → Next Layer → … → Output**

It is the core mechanism by which an MLP **turns raw input into meaningful predictions**.

---

##  Step-by-Step Explanation

1. **Input Layer**: Receives raw data.
    - Example: `[hours studied, sleep hours]`
2. **Hidden Layers**: Each performs:
    - Linear transform: `z = W*x + b`
    - Activation: `h = σ(z)`
    - Output `h` becomes input to the next layer
3. **Output Layer**: Produces the final prediction (e.g., class probability).

---



##  Beginner Analogy

Think of a **sandwich shop**:

- **Input** = raw ingredients (bread, veggies)
- **Linear transform** = layering ingredients in a specific way
- **Activation** = adding sauces to enhance flavor
- **Next layer** = pass sandwich to next chef for toppings
- **Output** = final sandwich ready to eat 🍴

---

##  Key Points

- Forward pass = **Input → Linear Transform → Activation → Next Layer → Output**
- Linear step → mixes inputs with weights + bias
- Activation → adds non-linearity to capture complex patterns
- Next layer → activated outputs become inputs for next layer
- Final output → network prediction

```python
import numpy as np

# --- Input ---
x = np.array([1, 0])  # input vector

# --- Hidden Layer ---
W1 = np.array([[0.2, 0.8],
               [0.6, 0.4]])   # weights for hidden layer
b1 = np.array([0.1, 0.2])    # bias for hidden layer

# Linear transformation
z1 = np.dot(W1, x) + b1

# ReLU activation
def relu(z):
    return np.maximum(0, z)

h1 = relu(z1)

print("Hidden layer output (after ReLU):", h1)

# --- Output Layer ---
W2 = np.array([0.5, 0.9])  # weights for output layer
b2 = 0.1                    # bias for output layer

# Linear transformation
z2 = np.dot(W2, h1) + b2

# Sigmoid activation
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

y = sigmoid(z2)

print("Final output (after Sigmoid):", y)
```

- `x` is the input vector with **2 features**: `[1, 0]`.
- These are the values we feed into the MLP.
- `W1` → weights connecting the **input layer to hidden layer** (2x2).
- `b1` → biases for the **hidden layer neurons**.
- `z1 = W1 * x + b1` → computes the **linear combination** of inputs for each hidden neuron:
    - Think of this as “mixing” the inputs with their importance (weights) and adding flexibility (bias).
- `relu(z1)` applies **ReLU activation** element-wise: `max(0, z)`
- ReLU introduces **non-linearity**, which allows the network to solve non-linear problems.
- `h1` → output of hidden layer neurons after activation.
- Applies **Sigmoid activation** on `z2`.
- Sigmoid **squashes the value between 0 and 1**, useful for probabilities (binary classification).
- `y` → final output of the network.
- Shows intermediate **hidden layer output**: `[0.3, 0.8]`
- Shows **final network prediction**: `0.73` → probability of class 1.

---
### ✅ Summary

This code is a **manual forward pass** of a tiny neural network with:

- 2 inputs → 2 hidden neurons (ReLU) → 1 output neuron (Sigmoid)
- Computes **step-by-step** the linear transformations and activations
- Outputs the **hidden layer values** and the **final prediction**

# 🧠 Loss Functions in Multi-Layer Perceptrons (MLPs)

A **loss function** is a mathematical measure of how wrong the MLP predictions are compared to the true targets.

It is **essential for training** because it tells the model **how to adjust its weights and biases** during backpropagation.

---

## 1️⃣ Mean Squared Error (MSE) – Regression

**✅ Purpose:**

- Used for predicting **continuous values** (e.g., house prices, temperature, stock prices).
- Measures the **average squared difference** between predicted and actual values.

**📐 Formula:**



**💡 Intuition:**

- Squares the error → larger mistakes are penalized more heavily.
- Encourages the model to reduce large errors.

**Example:**

- True value: 5, Prediction: 6 → Squared error = (6-5)² = 1
- Prediction: 8 → Squared error = (8-5)² = 9 → much larger penalty

---

## 2️⃣ Binary Cross-Entropy – Binary Classification

**✅ Purpose:**

- Used when output is **one of two classes** (e.g., spam or not spam).
- Treats output as **probability of class 1**.

**📐 Formula:**



**💡 Intuition:**

- Punishes confident but wrong predictions more than small mistakes.

**Example:**

- True: 1, Predicted: 0.9 → Loss ≈ 0.105 (small)
- True: 1, Predicted: 0.1 → Loss ≈ 2.3 (large)

---

## 3️⃣ Categorical Cross-Entropy – Multi-Class Classification

**✅ Purpose:**

- Used when each sample belongs to **exactly one class** and there are **more than 2 classes**.
- Works with **one-hot encoded targets**.

**📐 Formula:**



**💡 Intuition:**

- Measures dissimilarity between predicted probabilities and true class.
- Encourages network to assign high probability to correct class.

---

## 4️⃣ Sparse Categorical Cross-Entropy – Multi-Class with Integer Labels

**✅ Purpose:**

- Similar to categorical cross-entropy but works when **labels are integers** instead of one-hot vectors.
- Example: Instead of `[0,0,1,0]` for class 2, you just use `2`.

**📐 Formula:**

- Internally converts integer labels to one-hot, then applies categorical cross-entropy.

**Example:**

- True label = 2, Predicted probabilities = [0.1, 0.3, 0.6]



**💡 Advantage:**

- Simpler label representation → no need to manually one-hot encode targets.
- Works seamlessly with TensorFlow/Keras `sparse_categorical_crossentropy` loss.

---

## 5️⃣ Key Differences Between Classification Losses

| Loss Function | Task Type | Target Format | Output | Notes |
| --- | --- | --- | --- | --- |
| Binary Cross-Entropy | Binary Classification | 0 or 1 | Probability [0,1] | Only 2 classes |
| Categorical Cross-Entropy | Multi-Class Classification | One-hot vector | Probabilities | Each sample belongs to one class |
| Sparse Categorical Cross-Entropy | Multi-Class Classification | Integer labels | Probabilities | Converts labels internally; easier than one-hot |

---

## 6️⃣ How Loss Functions Work in MLPs

1. **Forward Pass:** Input → Hidden layers → Output
2. **Compute Loss:** Compare predictions with true labels using the chosen loss function
3. **Backpropagation:** Compute gradients of loss w.r.t. weights and biases
4. **Update Weights:** Use an optimizer (Adam, SGD) to reduce the loss

**🔑 Takeaway:**

- The choice of loss function directly affects learning.
- Using the wrong loss can prevent your MLP from learning properly.

```python
import numpy as np
import tensorflow as tf

# ----------------------------
# Sample Predictions & Targets
# ----------------------------

# Regression Example (MSE)
y_true_reg = np.array([3.0, 5.0, 2.5])
y_pred_reg = np.array([2.5, 5.0, 4.0])

mse = np.mean((y_true_reg - y_pred_reg)**2)
print("Mean Squared Error (MSE):", mse)

# ----------------------------
# Binary Classification Example
# ----------------------------
y_true_bin = np.array([1, 0, 1])
y_pred_bin = np.array([0.9, 0.2, 0.7])

binary_ce = -np.mean(y_true_bin*np.log(y_pred_bin) + (1-y_true_bin)*np.log(1-y_pred_bin))
print("Binary Cross-Entropy:", binary_ce)

# ----------------------------
# Multi-Class Classification (One-hot)
# ----------------------------
y_true_cat = np.array([[0, 0, 1], [1, 0, 0]])
y_pred_cat = np.array([[0.1, 0.2, 0.7], [0.8, 0.1, 0.1]])

cat_ce = -np.mean(np.sum(y_true_cat * np.log(y_pred_cat), axis=1))
print("Categorical Cross-Entropy (one-hot):", cat_ce)

# ----------------------------
# Sparse Categorical Cross-Entropy (Integer Labels)
# ----------------------------
y_true_sparse = np.array([2, 0])
y_pred_sparse = np.array([[0.1, 0.2, 0.7], [0.8, 0.1, 0.1]])

sparse_ce = tf.keras.losses.sparse_categorical_crossentropy(y_true_sparse, y_pred_sparse)
print("Sparse Categorical Cross-Entropy:", np.mean(sparse_ce.numpy()))
```

### ✅ What this code does:

- MSE – Calculates mean squared error for regression.
- Binary Cross-Entropy – Computes loss for binary classification.
- Categorical Cross-Entropy – For multi-class with one-hot targets.
- Sparse Categorical Cross-Entropy – For multi-class with integer labels (simpler).

# 🧠 Optimizers in Multi-Layer Perceptrons (MLPs)

In an MLP, optimizers control how the **weights and biases** of all layers are updated during training to minimize the loss function.  
Choosing the right optimizer can make learning **faster, more stable, and efficient**.

---

## 1️⃣ Stochastic Gradient Descent (SGD)

**How it works in MLPs:**
- Each layer has weights and biases.
- Forward pass: MLP computes outputs for inputs.
- Backward pass: Gradients of the loss w.r.t each weight are calculated.
- Update rule:


**Pros:**
- Simple and easy to implement
- Works well for small datasets

**Cons:**
- Can be slow on large datasets
- May get stuck in local minima or oscillate

**Example:**  
In a 2-layer MLP, SGD updates weights of hidden and output layers separately using their gradients.

---

## 2️⃣ Momentum

**How it works in MLPs:**
- Keeps a "memory" of previous weight updates.
- Accelerates updates in consistent gradient directions.
- Damps oscillations when gradients change sign.

**Formulas:**


**Pros:**
- Helps MLPs converge faster
- Smooths updates in layers with noisy gradients

**Example:**  
Hidden layer weights update considers both **current gradient** and **past updates** → faster learning of features.

---

## 3️⃣ Adam (Adaptive Moment Estimation)

**How it works in MLPs:**
- Combines momentum (first moment) and adaptive learning rates (second moment).
- Each weight gets a personalized learning rate based on gradient history.

**Formulas:**



**Pros:**
- Handles noisy gradients well
- Works well for deep MLPs with many layers
- Often requires less tuning of learning rate

**Example:**  
Each hidden and output layer weight is updated considering **current gradients**, **past gradients**, and **gradient magnitudes** → stable and fast learning.

---

## 🔑 Takeaways for MLPs

| Optimizer | Effect on MLP Training           | Pros                                | Cons                        |
|-----------|---------------------------------|------------------------------------|-----------------------------|
| SGD       | Updates weights layer by layer  | Simple, easy                        | Can be slow, oscillates     |
| Momentum  | Adds velocity to weight updates | Faster convergence, smoother        | Needs momentum hyperparameter|
| Adam      | Adaptive updates for each weight| Fast, stable, works for deep nets   | Slightly more computation   |

**Summary:**
- SGD is simple but may be slow.
- Momentum accelerates learning by remembering past updates.
- Adam is widely used for deep MLPs because it **adapts learning rates per weight** and is robust.


```python
#  MLP Optimizers Example on MNIST dataset
from tensorflow.keras.optimizers import SGD, Adam
# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0  # normalize pixel values
```

- Loads handwritten digits dataset (28x28 grayscale images, digits 0–9).
- Divides pixel values by 255 → normalizes them to [0,1], which helps the network learn faster.

```python
# Function to create a simple MLP
def create_mlp(activation=None):
    model = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(128, activation=activation),
        Dense(64, activation=activation),
        Dense(10, activation='softmax')
    ])
    return model
```

1. This function creates an MLP with:
- Flatten layer → converts 28x28 image into 784-element vector.
- Hidden layer 1 → 128 neurons, activation function passed as argument (e.g., ReLU).
- Hidden layer 2 → 64 neurons, activation function.
- Output layer → 10 neurons (for digits 0–9), softmax activation → outputs probabilities for each class.

```python
# 1️⃣ MLP with SGD

sgd_model = create_mlp(activation='relu')
sgd_model.compile(optimizer=SGD(learning_rate=0.01),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

print("Training with SGD...")
sgd_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2, verbose=2)
loss_sgd, acc_sgd = sgd_model.evaluate(x_test, y_test, verbose=0)
print("Test Accuracy (SGD): {:.2f}%".format(acc_sgd*100))
```

- Creates the MLP with ReLU activations.
- Optimizer: SGD (Stochastic Gradient Descent) with learning rate 0.01.
- Loss function: sparse_categorical_crossentropy (good for multi-class classification when labels are integers).
- Metrics: track accuracy
- Trains the model for 5 epochs.
- Batch size 128 → updates weights every 128 samples.
- Validation split 20% → monitors performance on unseen data.
- Evaluates the trained model on test data.
- Prints test accuracy for SGD.                                     

```python
# 2️⃣ MLP with Momentum

momentum_model = create_mlp(activation='relu')
momentum_model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9),
                       loss='sparse_categorical_crossentropy',
                       metrics=['accuracy'])

print("\nTraining with Momentum...")
momentum_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2, verbose=2)
loss_mom, acc_mom = momentum_model.evaluate(x_test, y_test, verbose=0)
print("Test Accuracy (Momentum): {:.2f}%".format(acc_mom*100))
```

- Same architecture as before.
- Adds momentum=0.9 to SGD → helps accelerate learning and smooth updates.
- Training and evaluation are similar to SGD

```python
# 3️⃣ MLP with Adam

adam_model = create_mlp(activation='relu')
adam_model.compile(optimizer=Adam(learning_rate=0.001),
                   loss='sparse_categorical_crossentropy',
                   metrics=['accuracy'])

print("\nTraining with Adam...")
adam_model.fit(x_train, y_train, epochs=5, batch_size=128, validation_split=0.2, verbose=2)
loss_adam, acc_adam = adam_model.evaluate(x_test, y_test, verbose=0)
print("Test Accuracy (Adam): {:.2f}%".format(acc_adam*100))
```

- Same MLP architecture.
- Optimizer: Adam → combines momentum + adaptive learning rates for faster, stable training.
- Learning rate smaller (0.001 is standard for Adam).
- Training and evaluation.
- Prints test accuracy for Adam

```python
#  Plotting Loss vs Epochs for Different Optimizers

import matplotlib.pyplot as plt

# Plot training loss for each optimizer

plt.figure(figsize=(10,6))

# SGD
plt.plot(sgd_model.history.history['loss'], label='SGD', marker='o')

# Momentum
plt.plot(momentum_model.history.history['loss'], label='Momentum', marker='x')

# Adam
plt.plot(adam_model.history.history['loss'], label='Adam', marker='s')

plt.title('Training Loss vs Epochs', fontsize=16)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

# 🧠 Weight Initialization in Multi-Layer Perceptrons (MLPs)

Proper weight initialization is **crucial** in MLPs for:

- **Preventing Vanishing/Exploding Gradients:** Ensures stable gradient flow during backpropagation.
- **Accelerating Convergence:** Leads to faster and more reliable training.
- **Avoiding Symmetry Breaking:** Starts neurons with distinct weights to learn different features.

---

## 1️⃣ Xavier (Glorot) Initialization

**Overview:**  
Proposed by Xavier Glorot and Yoshua Bengio, Xavier Initialization maintains consistent variance of activations across layers, ideal for **sigmoid** or **tanh** activations.




**When to Use:**  
- Layers with **sigmoid** or **tanh** activations.

**Importance:**  
- Prevents vanishing gradients.  
- Maintains activation variance across layers.  
- Ensures stable learning.

---

## 2️⃣ He Initialization

**Overview:**  
Introduced by Kaiming He et al., He Initialization is tailored for **ReLU** and its variants. It addresses the issue of **dying neurons** by accounting for ReLU’s non-linearity.




**When to Use:**  
- Layers with **ReLU**, **Leaky ReLU**, or **PReLU** activations.

**Importance:**  
- Prevents dying neurons.  
- Facilitates faster and stable convergence.

---

## ⚖️ Xavier vs. He Initialization

| Aspect                | Xavier Initialization        | He Initialization         |
|-----------------------|----------------------------|--------------------------|
| Activation Function    | Sigmoid, Tanh              | ReLU, Leaky ReLU, PReLU |
| Variance Scaling       | 2 / (fan_in + fan_out)     | 2 / fan_in               |
| Purpose                | Prevent vanishing/exploding gradients | Prevent dying neurons and accelerate convergence |
| Best For               | Shallow networks           | Deep ReLU-based networks |

---

## 🧪 Practical Impact in MLPs

- **Xavier Initialization:** Ensures stable gradient flow with sigmoid/tanh, avoids vanishing gradients.  
- **He Initialization:** Works with ReLU variants, prevents inactive neurons, and accelerates learning.

**Key Takeaway:**  
Choosing the right weight initialization is **vital** for MLPs to ensure **efficient learning**, **stable performance**, and **faster convergence**, especially in deep networks.

```python
import tensorflow as tf
from tensorflow.keras import layers, initializers

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Define MLP model
def build_mlp(initializer_type='xavier'):
    model = tf.keras.Sequential([
        layers.Flatten(input_shape=(28, 28)),
        layers.Dense(128, activation='relu', kernel_initializer=initializer_type),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

- Normalization (`/255.0`) scales pixel values to `[0,1]` for better neural network performance.
- `build_mlp()`: function to create a Multi-Layer Perceptron (MLP).
- **Layers:**
    - `Flatten`: Converts 28x28 image into a 784-element 1D vector.
    - `Dense(128, activation='relu')`: Hidden layer with 128 neurons, ReLU activation.
    - `Dense(10, activation='softmax')`: Output layer with 10 neurons (one per digit) and softmax activation for classification.

- **Compilation:**
    - `optimizer='adam'`: Adam optimizer for faster convergence.
    - `loss='sparse_categorical_crossentropy'`: suitable for multi-class classification with integer labels.
    - `metrics=['accuracy']`: tracks accuracy during training.


```python
# Initialize models
xavier_model = build_mlp(initializer_type=initializers.GlorotUniform())
he_model = build_mlp(initializer_type=initializers.HeUniform())

# Train models
xavier_model.fit(x_train, y_train, epochs=5, validation_split=0.2, verbose=1)
he_model.fit(x_train, y_train, epochs=5, validation_split=0.2, verbose=1)

# Evaluate models
xavier_loss, xavier_acc = xavier_model.evaluate(x_test, y_test)
he_loss, he_acc = he_model.evaluate(x_test, y_test)

print(f"Xavier Initialization - Test Accuracy: {xavier_acc*100:.2f}%")
print(f"He Initialization - Test Accuracy: {he_acc*100:.2f}%")
```

# 🧠 Batch vs Mini-Batch vs Online Training in MLPs

In Multi-Layer Perceptrons (MLPs), training involves updating network weights using gradient descent.  
The way we feed data to the network affects performance, convergence, and stability.  
This depends on the 'batch size', leading to three common training strategies: **Batch, Mini-Batch, and Online (Stochastic) Training**.

---

## 1️⃣ Batch Training 🏋️‍♂️

**📌 Definition:**  
- Updates weights after computing the gradient over the entire training dataset.

**⚙️ How it works:**  
1. Feed all training samples to the MLP at once.  
2. Compute the loss for all samples.  
3. Calculate gradients w.r.t all weights.  
4. Update the weights once per epoch.

**💡 Use case:**  
- Small datasets that fit into memory.  
- When stable gradient updates are required.

**👍 Pros:**  
- Accurate gradient estimate.  
- Smooth convergence.

**👎 Cons:**  
- Slow for large datasets.  
- Requires a lot of memory.

---

## 2️⃣ Mini-Batch Training 🏃‍♀️

**📌 Definition:**  
- Splits the dataset into smaller batches and updates weights after each mini-batch.

**⚙️ How it works:**  
1. Divide training data into batches (e.g., 32, 64, 128).  
2. Feed each mini-batch to the network.  
3. Compute loss and gradients for the mini-batch.  
4. Update weights after each mini-batch.

**💡 Use case:**  
- Large datasets.  
- Standard method in deep learning.

**👍 Pros:**  
- Balances memory usage and gradient accuracy.  
- Faster convergence than batch training.  
- Slight randomness helps escape local minima.

**👎 Cons:**  
- Some noise in gradients.  
- Batch size needs tuning.

---

## 3️⃣ Online (Stochastic) Training ⚡

**📌 Definition:**  
- Updates weights after each training example.

**⚙️ How it works:**  
1. Feed one sample to the MLP.  
2. Compute loss and gradient for that sample.  
3. Update weights immediately.  
4. Repeat for all samples.

**💡 Use case:**  
- Streaming data or very large datasets.  
- Immediate learning required.

**👍 Pros:**  
- Can start learning immediately.  
- Noisy updates can help escape local minima.

**👎 Cons:**  
- High variance in gradient → unstable convergence.  
- Requires smaller learning rate to stabilize.

---

## 🔍 Comparison Table 

| Feature                  | Batch                  | Mini-Batch           | Online                       |
|---------------------------|-----------------------|--------------------|-------------------------------|
| Weight Update Frequency   | Once per epoch        | After each mini-batch | After each training example  |
| Memory Requirement        | High (entire dataset) | Moderate (batch size dependent) | Low (one sample at a time) |
| Convergence Stability     | Very stable           | Moderate stability  | Noisy, may oscillate         |
| Training Speed            | Slow (large datasets) | Fastest for large datasets | Slow per epoch, but continuous |
| Gradient Accuracy         | Most accurate         | Moderate accuracy   | Noisy/approximate            |
| Use Case                  | Small datasets        | Most common in deep learning | Streaming data, real-time learning |
| MLP Application           | Shallow networks, small MNIST | Deep MLPs, large datasets | Online learning, adaptive systems |

---

## ✅ Key Takeaways 🎯

- **Batch 🏋️‍♂️**: Good for small datasets, stable convergence, memory-heavy.  
- **Mini-Batch 🏃‍♀️**: Standard for deep learning MLPs; balances speed, memory, and gradient accuracy.  
- **Online ⚡**: Good for streaming data; noisy updates can help escape local minima.


# 🧠 Layers in a Multi-Layer Perceptron (MLP)

An MLP consists of **three types of layers**: Input, Hidden, and Output. Each layer has a specific role in processing data and producing predictions.

---

## 1️⃣ Input Layer 🏁

📌 **Definition:**  
The first layer of the MLP that receives the raw input features from the dataset.

⚙️ **Characteristics:**  
- Number of neurons = number of input features.  
- No computation happens here besides passing the input forward.  

💡 **Example:**  
- For MNIST digit images of size 28×28, the input layer has 784 neurons (28×28 pixels flattened).  

---

## 2️⃣ Hidden Layers 🕵️‍♂️

📌 **Definition:**  
Layers between the input and output layers that perform **feature transformation** using weights, biases, and activation functions.

⚙️ **Characteristics:**  
- Can have one or more hidden layers.  
- Introduces **non-linearity** via activation functions (ReLU, Sigmoid, Tanh, etc.).  
- Learns **intermediate representations** of the input data.  

💡 **Example:**  
- An MLP with 2 hidden layers:  
  - 1st hidden layer: 128 neurons, ReLU activation  
  - 2nd hidden layer: 64 neurons, ReLU activation  

---

## 3️⃣ Output Layer 🎯

📌 **Definition:**  
The final layer that produces the network's predictions.

⚙️ **Characteristics:**  
- Number of neurons = number of output classes (for classification) or 1 (for regression).  
- Uses an **output activation function** appropriate for the task:  
  - Softmax → multi-class classification  
  - Sigmoid → binary classification  
  - Linear → regression  

💡 **Example:**  
- MNIST (10 digits) → Output layer has 10 neurons with softmax activation.  

---

## 🔍 Summary Table

|-----------------|----------------------------------------|---------------------------------------|-------------------------|
| Input Layer 🏁   | Receives raw input data                 | Equal to input features               | None                    |
| Hidden Layer 🕵️‍♂️ | Learns features & introduces non-linearity | Typically 32, 64, 128… (hyperparameter) | ReLU, Sigmoid, Tanh     |
| Output Layer 🎯  | Produces predictions                     | Equal to output classes (or 1)        | Softmax, Sigmoid, Linear |

# 🧠 Why ReLU is Preferred in Multi-Layer Perceptrons (MLPs)

## 1️⃣ ReLU Activation Function ⚡

The **Rectified Linear Unit (ReLU)** is defined as:


Where:  
- \(x\) = input to the neuron  
- Output = \(x\) if \(x > 0\), otherwise 0  

💡 Intuition: Passes positive values directly, blocks negative values.

---

## 2️⃣ Avoiding Vanishing Gradients 🌊

### Problem with Sigmoid/Tanh:
- Sigmoid/tanh saturate for large positive/negative inputs → derivatives become very small.  
- During backpropagation, gradients shrink layer by layer → **vanishing gradient problem**.  
- Makes deep MLPs **hard to train**.

### Why ReLU Helps:
- Gradient of ReLU = 1 for positive inputs.  
- Keeps gradient magnitude **constant** in active regions.  
- Allows **faster learning** and better signal propagation in deep networks.



---

## 3️⃣ Additional Advantages of ReLU ✅

- **Computationally efficient**: Simple max(0, x) operation.  
- **Sparse activation**: Only a subset of neurons fire → better representation.  
- **Works well with He Initialization**: Helps prevent dying neurons in deep MLPs.

---

## 4️⃣ Summary 🎯

| Feature                        | Sigmoid/Tanh             | ReLU                        |
|--------------------------------|-------------------------|-----------------------------|
| Gradient Magnitude             | Can vanish in deep layers | Constant for positive inputs |
| Computation                     | Expensive (sigmoid)      | Simple (max operation)       |
| Training Speed                  | Slow for deep MLPs       | Faster convergence           |
| Sparsity                        | Dense activations        | Sparse activations           |

✅ **Key Takeaway:** ReLU is preferred in deep MLPs because it **avoids vanishing gradients**, allows **faster training**, and is **computationally simple**.
